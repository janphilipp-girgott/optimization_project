```python
import pandas as pd
import os
```

```python
df = pd.read_csv("stock_ranking.csv")
df = df.drop(columns=["Name", "Last Sale", "Net Change", "% Change", "Country", "IPO Year", "Volume", "Sector", "Industry"])
df = df.sort_values(by="Market Cap", ascending=False).reset_index(drop=True)
```

```python
delete = pd.read_csv("to_be_deleted.csv")
delete = delete.drop(columns = ["Unnamed: 0"])
delete['Files to be deleted'] = delete['Files to be deleted'].apply(lambda x: os.path.basename(x).replace('.csv', ''))
```

```python
# Check if entries in a specific column of top500 are in the same column of delete
common_entries = df[df["Symbol"].isin(delete['Files to be deleted'])]
```

```python
# Remove entries in df that are common with delete
df = df[~df['Symbol'].isin(common_entries['Symbol'])].reset_index(drop=True)
df = df[~df['Symbol'].isin(["BRK/A", "BRK/B"])].reset_index(drop=True)
```

```python
top500 = df.iloc[0:500,0]
top500
```

```python
def create_stock_price_dataframe(stock_symbols, start_date, end_date):
    """
    Create a DataFrame with stock symbols as rows and dates as columns.
    Each cell contains the closing price of the stock on that date.

    :param stock_symbols: List of stock symbols
    :param start_date: Start date for the data (inclusive)
    :param end_date: End date for the data (inclusive)
    :return: DataFrame with stock prices
    """
    # Initialize an empty DataFrame to store stock prices
    stock_prices_df = pd.DataFrame()
    counter = []

    for symbol in stock_symbols:
        # Construct file path for each stock symbol
        file_path = f'stocks/{symbol}.csv'
        
        # Check if the file exists
        if os.path.exists(file_path):
            # Read the CSV file
            df = pd.read_csv(file_path)

            # Convert the 'Date' column to datetime for filtering
            df['Date'] = pd.to_datetime(df['Date'])

            # Filter the data for the given date range
            filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

            # If the DataFrame is empty, initialize it with the dates
            if stock_prices_df.empty:
                stock_prices_df = filtered_df[['Date']].set_index('Date')

            # Add the closing prices to the DataFrame
            stock_prices_df[symbol] = filtered_df.set_index('Date')['Close']
        else:
          counter.append(symbol) 

    stock_prices_df = stock_prices_df.transpose()

    return stock_prices_df, counter

# Define the stock symbols and date range
start_date = "2022-11-01"
end_date = "2023-10-31"

# Create the stock price DataFrame
actual_price_matrix_500, counter = create_stock_price_dataframe(top500, start_date, end_date)
```

```python
actual_price_matrix_500.to_csv("actual_price_matrix_500.csv")
actual_price_matrix_500
```
